```python
f(x) = cos(x)
f_integral = integrate(f, (x, -1, 1))
show(f_integral)
```

### Recursive functions in SageMath

Iteration in Sagemath

Consider a recursively defined iteration. 

### $x_{0}=4$

### $x_{k}=3x_{k-1}-10$ 

The “next” element in the sequence is equal to the previous times three and minus ten.

To calculate by hand takes a good deal of effort. 

### $x_{1}=3*4-10=2$ 

### $x_{2}=3*2-10=-4$ 

### $x_{3}=3*(-4)-10=-22$ 

What if we want to find $x_{50}$? We can use Sagemath to help us to evaluate $x_{50.}$

Defining a sequence recursively.  Suppose $x_{0}=4$ and $x_{k}=3x_{k-1}-10$ 

```python
# we define a function x
def x(k):
    if k==0:
        return 4
    else: 
        return 3*x(k-1)-10
```

```python
# Let's check our previous calculations 
show(x(1))
show(x(2))
show(x(3))

```

We can use this to find $x_{50.}$

```python
x(50)
```

 Wow!  That would have taken a long time to compute. 
 Now let's list out the first 50 iterates. 

```python
[(i,x(i)) for i in range(51)] #note we have to go up to 51 in order to see k = 50. 
```
